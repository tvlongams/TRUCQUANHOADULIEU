### Zachary's Karate Club 

```python
import networkx as nx
```

```python
G = nx.karate_club_graph()
```

```python
G.edges()
```

```python
adj_matrix_sparse = nx.adjacency_matrix(G)
adj_matrix_1 = adj_matrix_sparse.toarray()
print(adj_matrix_1)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6, 6))
ax = ax.imshow(adj_matrix_1, vmin=0, vmax=6)
plt.show()
```

```python
adj_matrix_sparse = nx.adjacency_matrix(G)
adj_matrix_2 = adj_matrix_sparse.toarray()
adj_matrix_2[adj_matrix_2> 0] = 1
print("Ma trận kề dạng 0, 1:")
print(adj_matrix_2)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6, 6))
ax = ax.imshow(adj_matrix_2, vmin=0, vmax=1)
plt.show()
```
